```python
# import
import pandas as pd
```

```python
# read in
df = pd.read_excel(r"C:\Users\user\Documents\DATA ANALYSIS\Pandas\Customer Call List.xlsx")
```

```python
# remove useless columns
df.drop(columns = 'Not_Useful_Column', inplace = True)
```

```python
# remove duplicates
df.drop_duplicates(inplace = True)
```

```python
# remove the trailling features for the last_name column
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
```

```python
# Standardize the phone number columns
# 1. remove the non-numericals in the phone numbers
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True)

# 2. convert the phone numbers to strings
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : str(x))

# 3. Standardize the phone numbers
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

#done
```

```python
# split the column
df[['Street', 'State', 'Zipcode']] = df['Address'].str.split(',', n = 2, expand = True)
```

```python
# standardize the paying customer and the do not contact columns
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
```

```python
df.fillna('')
```

```python
df.replace('nan--', '')
```

```python
# remove the rows which the customer doesnt want to be contacted
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)
```

```python
df.replace('nan--', '', inplace = True)
df.replace('Na--', '', inplace = True)
```

```python
# remove the rows where there is no numbers
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)
```
